```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 2, 3]
y2 = [2, 3, 4, 5]

# Plotting the lines
plt.plot(x, y1, label='Line 1', color='b', marker='o')  # Line 1 (x vs y1)
plt.plot(x, y2, label='Line 2', color='r', marker='x')  # Line 2 (x vs y2)

# Adding labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Plot of Two Lines')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 4]
y1 = [1, 4, 2, 3]
y2 = [2, 3, 4, 5]

# Plotting the lines with custom colors and markers
plt.plot(x, y1, label='Line 1', color='blue', marker='o', linestyle='-', linewidth=2, markersize=8)  # Blue line with circle markers
plt.plot(x, y2, label='Line 2', color='red', marker='x', linestyle='--', linewidth=2, markersize=8)  # Red dashed line with x markers

# Adding labels, title, and legend
plt.xlabel('X-axis')  
plt.ylabel('Y-axis')  
plt.title('Multiple Lines')  
plt.legend()  

# inorder to enable grid for better visibility
plt.grid(True)

plt.show()
```
